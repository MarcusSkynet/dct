```python
import numpy as np
import matplotlib.pyplot as plt

from reflection_wkb import R_wkb
```

```python
# ▒▒ 2.  User parameters ▒▒
mass_sol   = 30          # Black-hole mass in solar masses
ell        = 2           # Multipole index
fmin, fmax = 20, 2000    # Hz
Npoints    = 1500
freqs      = np.linspace(fmin, fmax, Npoints)
```

```python
# ▒▒ 3.  Compute reflection curve ▒▒
Rvals = R_wkb(freqs, l=ell, M_solar=mass_sol)
```

```python
# ▒▒ 4.  Simple plots ▒▒
fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(freqs, Rvals)
ax[0].set_xlabel("Frequency [Hz]"); ax[0].set_ylabel(f"R$_{{\\ell={ell}}}$")
ax[0].set_title("Linear-scale")

ax[1].semilogx(freqs, Rvals)
ax[1].set_xlabel("Frequency [Hz]")
ax[1].set_title("Log-x scale")

plt.tight_layout(); plt.show()
```
